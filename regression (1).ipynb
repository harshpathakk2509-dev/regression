Q6. What is the purpose of the least squares method in Simple Linear Regression?  
Ans: It minimizes the sum of squared differences between observed and predicted values, ensuring the best fit line.

Q7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?  
Ans: R² shows the proportion of variance in Y explained by X. Higher R² indicates better model fit.

Q8. What is Multiple Linear Regression?  
Ans: A regression technique that models the relationship between one dependent variable and two or more independent variables.

Q9. What is the main difference between Simple and Multiple Linear Regression?  
Ans: Simple regression uses one predictor, while multiple regression uses two or more predictors.

Q10. What are the key assumptions of Multiple Linear Regression?  
Ans: Linearity, independence, homoscedasticity, normality of residuals, and no multicollinearity among predictors.

Q11. What is heteroscedasticity, and how does it affect the results of a Multiple Linear Regression model?  
Ans: Heteroscedasticity occurs when residuals have non-constant variance. It makes estimates less reliable and standard errors biased.

Q12. How can you improve a Multiple Linear Regression model with high multicollinearity?  
Ans: Use techniques like removing correlated variables, applying Principal Component Analysis (PCA), or using regularization methods (Ridge/Lasso).

Q13. What are some common techniques for transforming categorical variables for use in regression models?  
Ans: One-hot encoding, label encoding, and dummy variables.

Q14. What is the role of interaction terms in Multiple Linear Regression?  
Ans: Interaction terms capture the combined effect of two variables on the dependent variable.

Q15. How can the interpretation of intercept differ between Simple and Multiple Linear Regression?  
Ans: In simple regression, the intercept is Y when X=0. In multiple regression, it is Y when all predictors equal zero, which may not always be meaningful.

Q16. What is the significance of the slope in regression analysis, and how does it affect predictions?  
Ans: The slope indicates the rate of change in Y for each unit change in X. It directly affects predictions.

Q17. How does the intercept in a regression model provide context for the relationship between variables?  
Ans: It sets the baseline value of Y, helping interpret the effect of predictors relative to that baseline.

Q18. What are the limitations of using R² as a sole measure of model performance?  
Ans: R² does not account for overfitting, does not indicate causation, and ignores model complexity.

Q19. How would you interpret a large standard error for a regression coefficient?  
Ans: It means the coefficient estimate is unstable and less reliable, possibly due to multicollinearity or insufficient data.

Q20. How can heteroscedasticity be identified in residual plots, and why is it important to address it?  
Ans: Residual plots show patterns like increasing spread. Addressing it ensures valid statistical inference.

Q21. What does it mean if a Multiple Linear Regression model has a high R² but low adjusted R²?  
Ans: It suggests overfitting — the model explains variance but adds unnecessary predictors that don’t improve performance.

Q22. Why is it important to scale variables in Multiple Linear Regression?  
Ans: Scaling ensures predictors are on comparable scales, improving model stability and interpretation, especially in regularization.

Q23. What is polynomial regression?  
Ans: A regression technique where the relationship between X and Y is modeled using polynomial terms of X.

Q24. How does polynomial regression differ from linear regression?  
Ans: Linear regression fits a straight line, while polynomial regression fits a curved line by including higher-order terms.

Q25. When is polynomial regression used?  
Ans: When the relationship between variables is nonlinear but can be approximated by a polynomial curve.

Ans: Yes, by including polynomial terms for each predictor and their interactions.

Q28. What are the limitations of polynomial regression?  
Ans: Risk of overfitting, poor extrapolation, and increased complexity with higher degrees.

Q29. What methods can be used to evaluate model fit when selecting the degree of a polynomial?  
Ans: Cross-validation, adjusted R², AIC/BIC, and residual analysis.

Q30. Why is visualization important in polynomial regression?  
Ans: Visualization helps understand the curve fit, detect overfitting, and interpret relationships.
