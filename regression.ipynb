Q2. What are the key assumptions of Simple Linear Regression?  
Ans:

Linearity: The relationship between X and Y is linear.

Independence: Observations are independent of each other.

Homoscedasticity: Constant variance of residuals across values of X.

Normality: Residuals are normally distributed.

No multicollinearity (only one predictor in simple regression).

Q6. What is the purpose of the least squares method in Simple Linear Regression?  
Ans: The least squares method minimizes the sum of squared differences between observed values and predicted values. This ensures the regression line fits the data as closely as possible.

Q7. How is the coefficient of determination (R²) interpreted in Simple Linear Regression?  
Ans: R² measures the proportion of variance in the dependent variable explained by the independent variable.

R² = 0 → No explanatory power.

R² = 1 → Perfect fit.
Higher R² indicates a better model fit.
